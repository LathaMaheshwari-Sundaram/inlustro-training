```python
import seaborn as sns
```

```python
plt.figure(figsize=(5,4))
sns.boxplot(data=pd[["Age","Fare"]],palette="muted")
plt.title("Boxplot for outlier detection")
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x="Sex",data=pd,palette="colorblind")
plt.title("Sex count",fontsize=18)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x="Sex",y="Survived",data=pd,palette="deep")
plt.title("Survival rate by sex",fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(pd["Age"],bins=25,kde=True,palette="mako")
plt.title("Age distribution",fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(data=pd,x="Age",y="Fare",palette="dark")
plt.title("Scatterplot",fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.kdeplot(x="Age",data=pd)
plt.title("density plot")
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.violinplot(x="Age",y="Sex",data=pd)
plt.title("violin plot",fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.lineplot(x="Age",y="Fare",data=pd)
plt.title("linechart",fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(5,4))
plt.pie(x="Fare",data=pd)
plt.title("piechart",fontsize=16)
plt.show()
```
